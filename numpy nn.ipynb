```python
import numpy as np
```

```python
x = np.array([
        [ 1, 0 ],
        [0, 1],
        [1, 1],
        [0, 0],
]
)
y = np.array([
    [0],
    [0],
    [1],
[0]]
)
lr = 0.01
```

```python
def sigmoid(x):
    return 1/(1+(np.exp(-x)))
def sigp(x):
    return x * (1 - x)
```

```python
inp = 2
n0 = 10
w0 = np.random.random(inp * n0).reshape((n0, inp))
b0 = np.zeros(n0)
n1 = 1
w1 = np.random.random(n0 * n1).reshape((n1, n0))
b1 = np.zeros(n1)
print("w0", w0)
print("b0", b0)
print()
print("w1", w1)
print("b1", b1)
```

```python
for i in range(1000000):
#     print("Hidden layer 1:")
    bout0 = np.dot(x , w0.T) + b0
#     print("bout0", bout0)
    out0 = np.apply_along_axis(sigmoid, axis = 0, arr = bout0)
#     print("out0", out0)

#     print("\nHidden layer 2:")
    bout1 = np.dot(out0 , w1.T) + b1
#     print("bout1", bout1)
    out1 = np.apply_along_axis(sigmoid, axis = 0, arr = bout1)
    print("out1 {}".format("".join(list(map(lambda x:" 0{:.6f}".format(x[0]),out1)))), end="\r")


    gmse = out1 - y
#     print('mse grad', gmse)

    gout1 = gmse * np.apply_along_axis(sigp, axis=0, arr = out1)
#     print("out1 grad", gout1)
    b1 -= lr * np.sum(gout1, axis=0)
    w1 -= lr * np.dot(out0.T, gout1).T
    gout0 = np.dot(gout1 , w1) * np.apply_along_axis(sigp, axis=0, arr=out0)
#     print("out0 grad", gout0)
    b0 -= lr * np.sum(gout0, axis=0)
    w0 -= lr * np.dot(x.T, gout0).T
```
